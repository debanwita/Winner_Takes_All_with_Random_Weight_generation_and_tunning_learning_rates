```python
import numpy as np
```

```python
class competitive_learning(object):
    def __init__(self):
        
        self.Points = np.array(([-1, 1],
                               [-1, 2],
                               [1, 0],
                               [1, 2],
                               [2, -1],
                               [2, 2]), dtype=float)
        self.Clusters =np.array([1, 1, 2, 0, 2, 0],dtype=int)
        self.new_cluster=np.array([0, 0, 0, 0, 0, 0], dtype=int)
       
    def cl(self):
        
        #self.W = np.random.rand(2, 3)
        j = 0
        for rate in np.arange(0.01, 0.1, 0.01):
            print("Learning rate :" , rate)
            print("-------------------------")
            self.W = np.array(([0, -1, 2], [2, 0, 1]), dtype=float)
            j = 0
            while(1):
                flag = 0 
                print("\tIteration-", j)
                print("============================")
                print("No.", "\t", "Points", "\t", "D0", "\t", "D1", "\t", "D2", "\t", "[W00,W10]", "\t", "[W01,W11]", "\t", "[W02,W12]", "\t", "Cluster_no" )
                print("----------------------------------------------------------------------------------------------------------------------------")
                for i in range(np.size(self.Points, 0)):
                    d0 = np.square(self.W[0, 0] - self.Points[i, 0]) + np.square(self.W[1, 0] - self.Points[i, 1])

                    d1 = np.square(self.W[0, 1] - self.Points[i, 0]) + np.square(self.W[1, 1] - self.Points[i, 1])


                    d2 = np.square(self.W[0, 2] - self.Points[i, 0]) + np.square(self.W[1, 2] - self.Points[i, 1])
                   
                    if(d0<d1 and d0<d2):
                        #print("cluster0")

                        self.W[0, 0] = self.W[0, 0] + (rate * (self.Points[i, 0] - self.W[0, 0]))
                        self.W[1, 0] = self.W[1, 0] + (rate * (self.Points[i, 1] - self.W[1, 0]))

                        self.new_cluster[i] = 0

                        #print(i, "\t", self.Points[i], "\t", d0, "\t", d1, "\t", d2, "\t", self.W[0,0], ", " self.W[1,0] , "\t", self.W[0,1], "," , self.W[1, 1] , "\t", self.W[0, 2], ",", self.W[1, 2], "\t", "0")
                        print(i, "\t", self.Points[i], "\t", round(d0, 1), "\t", round(d1, 1), "\t", round(d2, 1), "\t", round(self.W[0][0], 2), ";", round(self.W[1][0], 2), "\t", round(self.W[0][1], 2), ";", round(self.W[1][1], 2), "\t", round(self.W[0][2], 2), ";", round(self.W[1][2], 2), "\t", "0");

                    if(d1<d0 and d1<d2):
                        #print("cluster1")

                        self.W[0, 1] = self.W[0, 1] + (rate * (self.Points[i, 0] - self.W[0, 1]))
                        self.W[1, 1] = self.W[1, 1] + (rate * (self.Points[i, 1] - self.W[1, 1]))

                        #print(i, "\t", self.Points[i], "\t", d0, "\t", d1, "\t", d2, "\t", self.W[0, 0], ", " self.W[1, 0] , "\t", self.W[0,1], "," , self.W[1, 1] , "\t", self.W[0, 2], ",", self.W[1, 2], "\t", "1")
                        print(i, "\t", self.Points[i], "\t", round(d0,1), "\t", round(d1, 1), "\t", round(d2, 1) , "\t", round(self.W[0][0], 2), ";",  round(self.W[1][0], 2), "\t", round(self.W[0][1], 2), ";", round(self.W[1][1], 2), "\t", round(self.W[0][2], 2), ";", round(self.W[1][2], 2), "\t", "1");
                        self.new_cluster[i] = 1

                    if(d2<d0 and d2<d1):
                        #print("cluster2")

                        self.W[0, 2] = self.W[0, 2] + (rate * (self.Points[i, 0] - self.W[0, 2]))
                        self.W[1, 2] = self.W[1, 2] + (rate * (self.Points[i, 1] - self.W[1, 2]))

                        #print(i, "\t", self.Points[i], "\t", d0, "\t", d1, "\t", d2, "\t", self.W[0,0], ", " self.W[1,0] , "\t", self.W[0,1], "," , self.W[1, 1] , "\t", self.W[0, 2], ",", self.W[1, 2], "\t", "2")
                        print(i, "\t", self.Points[i], "\t", round(d0, 1), "\t", round(d1, 1), "\t", round(d2, 1), "\t", round(self.W[0][0],2), ";", round(self.W[1][0],2), "\t", round(self.W[0][1], 2), ";", round(self.W[1][1], 2), "\t", round(self.W[0][2], 2), ";", round(self.W[1][2], 2), "\t", "2");

                        self.new_cluster[i] = 2

                    #print(self.W)
                    #print("-----")
                    print()


                j = j + 1
               # print("clusters", self.Clusters)
                #print("new cluster", self.new_cluster)
                for i in range(np.size(self.Clusters, 0)):
                    if(self.Clusters[i] != self.new_cluster[i]):
                        flag = 1
                        break
                if(flag == 0):
                    break

               

            
        
```
